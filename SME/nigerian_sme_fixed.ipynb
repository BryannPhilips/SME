```python
numeric_cols = df.select_dtypes(include=['int64','float64']).columns

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    
    print(f"{col}: {len(outliers)} outliers")
```

## Outlier Treatment:

##### Outliers were examined using the Interquartile Range (IQR) method. While some extreme values were detected, they were retained because they likely reflect genuine variations in SME sales performance rather than data entry errors. Removing them could distort real-world business patterns, so the dataset was preserved in its original form.

```python
import matplotlib.ticker as mtick  # Added import for mtick

plt.figure(figsize=(10,5))
sns.histplot(df['monthly_sales'], kde=True, color='green')
fmt = '₦{x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
plt.gca().xaxis.set_major_formatter(tick)
plt.title("Distribution of Monthly Sales (Naira)")
plt.xlabel("Monthly Sales (Naira)")
plt.ylabel("Frequency")
plt.show()
```

##### The distribution of monthly sales is right-skewed, meaning that while the majority of SMEs earn within a specific mid-range (the peak of the curve), there are fewer businesses achieving very high sales volumes. This concentration of data suggests that most SMEs face similar market constraints, while the "long tail" represents top-performing outliers who have successfully optimized factors like location or digital presence to significantly outperform the average.

```python
import matplotlib.ticker as mtick

plt.figure(figsize=(12,6))

sns.boxplot(
    x='business_type',
    y='monthly_sales',
    hue='business_type',   # fixes the warning
    data=df,
    palette='Set2',
    legend=False
)

plt.xticks(rotation=45)
plt.title("Monthly Sales by Business Type")

# Format axis to Naira
plt.gca().yaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, pos: f'₦{int(x):,}')
)

plt.ylabel("Monthly Sales (Naira)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

plt.figure(figsize=(12,6))

avg_sales = df.groupby('business_type')['monthly_sales'].mean().sort_values()

colors = ['#4C72B0', '#55A868'] * (len(avg_sales)//2 + 1)

bars = plt.bar(avg_sales.index, avg_sales.values, color=colors[:len(avg_sales)])

plt.xticks(rotation=45)
plt.title("Average Monthly Sales by Business Type")

plt.gca().yaxis.set_major_formatter(
    mtick.FuncFormatter(lambda x, pos: f'₦{int(x):,}')
)

for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        yval,
        f'₦{int(yval):,}',
        ha='center',
        va='bottom',
        fontsize=9
    )

plt.ylabel("Average Monthly Sales (Naira)")
plt.show()
```

##### Average monthly sales vary across business types, indicating that the nature of the business significantly influences revenue potential.

#### Following the evaluation of multiple regression models, LightGBM emerged as the best-performing algorithm for predicting monthly SME sales.

#### The model achieved an R² of 0.8684, indicating that approximately 87% of the variation in sales is explained by the predictor variables. This reflects a strong model fit and suggests that the selected features effectively capture the primary factors influencing business revenue.

#### Additionally, the model recorded the lowest error metrics among all candidates (MAE: 2030.60; RMSE: 2794.62), demonstrating superior predictive accuracy and consistency.

#### Although alternative models such as Gradient Boosting Regressor produced competitive results, their slightly lower R² values indicate reduced explanatory power. The stronger performance of LightGBM suggests that the relationship between the predictors and monthly sales is likely nonlinear, making ensemble-based methods more suitable than traditional regression approaches.

#### Overall, the results confirm that LightGBM provides the most reliable framework for forecasting SME sales within this dataset and was therefore selected as the final model.

```python
tuned_model = tune_model(best_model)
```

#### Hyperparameter tuning was conducted on the selected LightGBM model to improve predictive performance. However, the tuned version produced a slightly lower R² of 0.8632 compared to the original model (0.8684).

#### Importantly, this outcome does not weaken the modeling process. Instead, it demonstrates proper machine learning practice — validating whether optimization genuinely improves performance rather than assuming it will.

#### As a result, the original LightGBM model was retained as the final model because it offers stronger explanatory power and lower prediction error.
